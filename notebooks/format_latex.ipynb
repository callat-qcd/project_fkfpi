### Make table

```python
output = '''\
\\begin{ruledtabular} \\centering
\\begin{tabular}{ l c c c c}
Model & $F_K/F_\pi$ & logGBF & $\chi_\\nu$ & $Q$ \\\ \\midrule[0.5pt]
'''

for j, model in enumerate(sorted(list(fit_results))):
    if j%5 == 0:
        output += '\\midrule[0.5pt] \n'
        
    results = [fit_results[model][key] for key in ['name', 'FK/Fpi', 'logGBF', 'chi2/df', 'Q']]
    output += '{} & {} & {:.2f} & {:.2f} & {:.2f} \\\ \n'.format(*results).replace('_', '\_')


output += '''\
\\end{tabular}
\\end{ruledtabular}
'''


print(output)
```
